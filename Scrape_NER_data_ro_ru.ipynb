## Scrape individual pages that contain lost/found animals

```python
import numpy as np

def analyze_individual_page(link, language='ro'):
    
    
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')

#     print(page.content)
    
    html = list(soup.children)[2]
    body = list(html.children)[3]
#     print(body)

    divs = body.find_all("div")
#     print(body.prettify())

    for item in divs:
        if 'itemprop' in item:
            print(item)
    
#     for description in body.find_all(class_='adPage__content__description'):
#         print(description.text)
    
    description = body.find(class_='adPage__content__description')
    if description:
        print(description.text)
    
    region_el = body.find(class_='adPage__content__region')
    location = region_el.text.replace("Регион:", '').strip().replace("  ", "")
    print("--location:--", location)
    
    phone_el = body.find(class_='adPage__content__phone')
    phone = None
    if phone_el:
        phone =  phone_el.text.replace("Контакты:", '').strip()
        print("--phone:--", phone)
    
    d = {
        "link": [link] if link else [np.nan],
        "text": [description.text] if description else [np.nan],
        "location": [location] if location else [np.nan],
        "phone": [phone] if phone else [np.nan],
        "language": [language] if language else [np.nan]
    }
    #df = pd.DataFrame(columns=["link", "text", "location", "phone", "language"])
    df = pd.DataFrame(data=d)
    
    return df
    
    
analyze_individual_page('https://999.md/ru/69334577')
```

```python
import requests

df_res = pd.DataFrame(columns=["link", "text", "location", "phone", "language"])

for link, title in links_lost_found_animals_ro.items():
    print(title)
    
    full_link = 'https://999.md' + link
    print(full_link)
    df1 = analyze_individual_page(full_link)
    df_res = df_res.append(df1)    

    print("-----")
```
