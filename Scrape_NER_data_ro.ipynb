```python
links_lost_found_animals_ro = {}
links_lost_found_animals_ru = {}

words_to_search_ru = ['пропал', 'нашел', 'убежал', 'нашё', 'нашлась']
words_to_search_ro = ['pierdut', 'gasit', 'găsit']

for item in soup.find_all(class_="ads-list-photo-item"):
    item_title_div = item.find_all(class_="ads-list-photo-item-title")[0]
    item_title = item_title_div.find_all("a")[0]
    print(item_title.text)
    print('Link:' + 'https://999.md/' + item_title['href'])
    
    
    for word in words_to_search_ro:
        if word in item_title.text.lower():
            links_lost_found_animals_ro[item_title['href']] = item_title.text
    
    for word in words_to_search_ru:
        if word in item_title.text.lower():
            links_lost_found_animals_ru[item_title['href']] = item_title.text
    
#     print(item_title)
```

## Scrape individual pages that contain lost/found animals

```python
def analyze_individual_page(link):
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')

#     print(page.content)
    
    html = list(soup.children)[2]
    body = list(html.children)[3]
#     print(body)

    divs = body.find_all("div")
#     print(body.prettify())

    for item in divs:
        if 'itemprop' in item:
            print(item)
    
    for description in body.find_all(class_='adPage__content__description'):
        print(description.text)
    
    region = body.find(class_='adPage__content__region')
    print("--location:--", region.text.replace("Регион:", '').strip().replace("  ", ""))
    
    phone = body.find(class_='adPage__content__phone')
    print("--phone:--", phone.text.replace("Контакты:", '').strip())
    
        
    
    
analyze_individual_page('https://999.md/ru/69334577')
```

```python
import requests

link = 'https://www.animalepierdute.ro/caini-si-pisici-pierdute/'
link = 'https://999.md/ru/list/animals-and-plants/missing'
# link = "http://dataquestio.github.io/web-scraping-pages/simple.html"

for link, title in links_lost_found_animals_ro.items():
    print(title)
    
    full_link = 'https://999.md' + link
    print(full_link)
    analyze_individual_page(full_link)
    
    

    print("-----")
```
